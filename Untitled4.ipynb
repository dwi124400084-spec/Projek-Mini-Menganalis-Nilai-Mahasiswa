```python
# Membaca Dataset
import csv

def baca_dataset(filepath):
    data = []
    with open(filepath, "r", encoding="utf-8") as f:
        reader = csv.reader(f, delimiter=None, skipinitialspace=True)

        header = next(reader)          # baca nama kolom
        header = [h.strip() for h in header]

        for row in reader:
            if not row or all(x.strip() == "" for x in row):
                continue   # skip baris kosong
            row = [x.strip() for x in row]
            record = dict(zip(header, row))
            data.append({
                "NIM": record["NIM"],
                "Nama": record["Nama"],
                "Tugas": int(record["Tugas"]),
                "UTS": int(record["UTS"]),
                "UAS": int(record["UAS"])
            })

    return data
```

```python
# Class Mahasiswa OOP
# Class Mahasiswa OOP
class Mahasiswa:
    def __init__(self, nim, nama, tugas, uts, uas):
        self.nim = nim
        self.nama = nama
        self.tugas = tugas
        self.uts = uts
        self.uas = uas

    def nilai_akhir(self):
        return 0.3*self.tugas + 0.3*self.uts + 0.4*self.uas

    def kategori(self):
        n = self.nilai_akhir()
        if n >= 80: return "A"
        elif n >= 70: return "B"
        elif n >= 60: return "C"
        elif n >= 50: return "D"
        return "E"
```

```python
# Mengubah Dataset ke Objek
obj_mahasiswa = []

for d in data:
    m = Mahasiswa(
        d["NIM"],
        d["Nama"],
        d["Tugas"],
        d["UTS"],
        d["UAS"]
    )
    obj_mahasiswa.append(m)
```

```python
# Fungsi Analisis
def rata_rata(obj):
    return sum([m.nilai_akhir() for m in obj]) / len(obj)

def cari_tertinggi(obj):
    return max(obj, key=lambda x: x.nilai_akhir())

def cari_terendah(obj):
    return min(obj, key=lambda x: x.nilai_akhir())
```

```python
# Menyimpan Hasil Analisis ke TXT
output_path = "hasil_analisis.txt"

with open(output_path, "w", encoding="utf-8") as f:
    f.write("Hasil Analisis Nilai Mahasiswa\n")
    f.write("----------------------------------\n\n")

    jumlah = len(obj_mahasiswa)
    f.write(f"Jumlah Mahasiswa: {jumlah}\n")

    if jumlah == 0:
        f.write("Rata-rata nilai akhir: - (data kosong)\n")
        f.write("Nilai Tertinggi: -\n")
        f.write("Nilai Terendah: -\n")
    else:
        f.write(f"Rata-rata nilai akhir: {rata_rata(obj_mahasiswa):.2f}\n")
        tertinggi = cari_tertinggi(obj_mahasiswa)
        terendah = cari_terendah(obj_mahasiswa)
        f.write(f"Nilai Tertinggi: {tertinggi.nama} ({tertinggi.nilai_akhir():.2f})\n")
        f.write(f"Nilai Terendah: {terendah.nama} ({terendah.nilai_akhir():.2f})\n")
print("✔ File berhasil disimpan:", output_path)
```

```python
import os
from openpyxl import Workbook

# ====== DATA MAHASISWA ======
data = [
    ("124400084","Maharani",78,42,52),
    ("124400102","Alya Septia",68,54,54),
    ("124400034","Rachmat Fadilah",54,23,65),
    ("124400088","Agam Wahyu",94,46,43),
    ("124400082","Aurelya Ibra",86,52,67),
    ("124400022","Athira Irany",81,76,54),
    ("124400038","Alfahri Dhika",72,82,48),
    ("124400072","Ryan Saputra",48,32,56),
    ("124400085","Justiana",68,45,26),
    ("124400081","Ganendra",66,78,35),
    ("124400073","Hisyam Musfeer",45,44,24),
]

# ====== FUNGSI NILAI ======
def nilai_akhir(tugas, uts, uas):
    return 0.3*tugas + 0.3*uts + 0.4*uas

def kategori(n):
    if n >= 80: return "A"
    elif n >= 70: return "B"
    elif n >= 60: return "C"
    elif n >= 50: return "D"
    return "E"

# ====== NAMA FOLDER & FILE ======
folder = "output_excel"
filename = "Nilai_Mahasiswa.xlsx"

# Buat folder otomatis jika belum ada
if not os.path.exists(folder):
    os.makedirs(folder)

full_path = os.path.join(folder, filename)

# ====== BUAT EXCEL ======
wb = Workbook()
ws = wb.active
ws.title = "Nilai Mahasiswa"

# Header
ws.append(["NIM", "Nama", "Tugas", "UTS", "UAS", "Nilai Akhir", "Kategori"])

# Data
for nim, nama, tugas, uts, uas in data:
    na = nilai_akhir(tugas, uts, uas)
    kat = kategori(na)
    ws.append([nim, nama, tugas, uts, uas, round(na,2), kat])

# Simpan otomatis
wb.save(full_path)

print("✔ File Excel otomatis tersimpan di:", full_path)
```
